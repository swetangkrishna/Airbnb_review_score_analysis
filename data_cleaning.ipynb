```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


plt.style.use('ggplot')
plt.rcParams['lines.linewidth'] = 1.7
plt.rcParams["figure.figsize"] = (10,6)
```

```python
w = [1,2,3,4]

df1 = pd.Series(w, index=['a','b','c','d'], dtype = 'float',name = 'panda')
print(df1)
```

```python
dic = {'a' : '1',
       'b' : '2',
       'c' : '3',
       }

df2 = pd.Series(dic, name = 'dictionary')
print(df2)
```

```python
dict = {"name" : ['pyhton', 'java','c++'] , 'por' : ['10','20','30'], 'rank':[1,2,3] }
df3 = pd.Series(dict, name = "dictionary")
print(df3)
```
